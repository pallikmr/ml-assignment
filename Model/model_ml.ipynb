```python
#Splitting data into training and testing
X = df.drop('quality', axis=1)
y = df['quality']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
from sklearn.metrics import (
    accuracy_score,
    roc_auc_score,
    precision_score,
    recall_score,
    f1_score,
    matthews_corrcoef
)
```

```python
def evaluate_model(model, X_test, y_test, y_pred):
    results = {}
    results['Accuracy'] = accuracy_score(y_test, y_pred)
    results['AUC'] = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])
    results['Precision'] = precision_score(y_test, y_pred)
    results['Recall'] = recall_score(y_test, y_pred)
    results['F1 Score'] = f1_score(y_test, y_pred)
    results['MCC'] = matthews_corrcoef(y_test, y_pred)
    return results
```

```python
#1.Logistic Regression
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()
model1.fit(X_train, y_train)
```

```python
y_pred = model1.predict(X_test)
```

```python
lr_results = evaluate_model(model1, X_test, y_test, y_pred)
print("Logistic Regression:", lr_results)
```

```python
#2.Desicion Tree Classifier
from sklearn.tree import DecisionTreeClassifier
model2 = DecisionTreeClassifier(random_state=42)
model2.fit(X_train, y_train)
```

```python
y_pred = model2.predict(X_test)
```

```python
dt_results = evaluate_model(model2, X_test, y_test, y_pred)
print("Decision Tree:", dt_results)
```

```python
#3.KNN
from sklearn.neighbors import KNeighborsClassifier
model3 = KNeighborsClassifier(n_neighbors=5)
model3.fit(X_train, y_train)
```

```python
y_pred = model3.predict(X_test)
```

```python
knn_results = evaluate_model(model3, X_test, y_test, y_pred)
print("KNN:", knn_results)
```

```python
#4.Naive Bayes Classifier
from sklearn.naive_bayes import GaussianNB
model4 = GaussianNB()
model4.fit(X_train, y_train)
```

```python
y_pred = model4.predict(X_test)
```

```python
nb_results = evaluate_model(model4, X_test, y_test, y_pred)
print("Naive Bayes:", nb_results)
```

```python
#5.Ensemble Model- Random Forest
from sklearn.ensemble import RandomForestClassifier
model5= RandomForestClassifier(random_state=42)
model5.fit(X_train, y_train)
```

```python
y_pred = model5.predict(X_test)
```

```python
rf_results = evaluate_model(model5, X_test, y_test, y_pred)
print("Random Forest:", rf_results)
```
